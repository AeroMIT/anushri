```python
import cv2
import numpy as np

def nothing(x):
    pass

# Load the image
image = cv2.imread(r"C:\Users\ANUSHRI\Downloads\fruits opencv.jpeg")  # Replace with your image path
image = cv2.resize(image, (600, 400))  # Resize for better display

# Convert image to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Create a window for trackbars
cv2.namedWindow("Trackbars")

# Create trackbars for the first red range
cv2.createTrackbar("Lower H1", "Trackbars", 0, 180, nothing)
cv2.createTrackbar("Lower S1", "Trackbars", 100, 255, nothing)
cv2.createTrackbar("Lower V1", "Trackbars", 100, 255, nothing)
cv2.createTrackbar("Upper H1", "Trackbars", 10, 180, nothing)
cv2.createTrackbar("Upper S1", "Trackbars", 255, 255, nothing)
cv2.createTrackbar("Upper V1", "Trackbars", 255, 255, nothing)

# Create trackbars for the second red range
cv2.createTrackbar("Lower H2", "Trackbars", 170, 180, nothing)
cv2.createTrackbar("Lower S2", "Trackbars", 100, 255, nothing)
cv2.createTrackbar("Lower V2", "Trackbars", 100, 255, nothing)
cv2.createTrackbar("Upper H2", "Trackbars", 180, 180, nothing)
cv2.createTrackbar("Upper S2", "Trackbars", 255, 255, nothing)
cv2.createTrackbar("Upper V2", "Trackbars", 255, 255, nothing)

while True:
    # Get trackbar positions for first red range
    l_h1 = cv2.getTrackbarPos("Lower H1", "Trackbars")
    l_s1 = cv2.getTrackbarPos("Lower S1", "Trackbars")
    l_v1 = cv2.getTrackbarPos("Lower V1", "Trackbars")
    u_h1 = cv2.getTrackbarPos("Upper H1", "Trackbars")
    u_s1 = cv2.getTrackbarPos("Upper S1", "Trackbars")
    u_v1 = cv2.getTrackbarPos("Upper V1", "Trackbars")

    # Get trackbar positions for second red range
    l_h2 = cv2.getTrackbarPos("Lower H2", "Trackbars")
    l_s2 = cv2.getTrackbarPos("Lower S2", "Trackbars")
    l_v2 = cv2.getTrackbarPos("Lower V2", "Trackbars")
    u_h2 = cv2.getTrackbarPos("Upper H2", "Trackbars")
    u_s2 = cv2.getTrackbarPos("Upper S2", "Trackbars")
    u_v2 = cv2.getTrackbarPos("Upper V2", "Trackbars")

    # Define HSV ranges for red (two ranges)
    lower_red1 = np.array([l_h1, l_s1, l_v1])
    upper_red1 = np.array([u_h1, u_s1, u_v1])
    lower_red2 = np.array([l_h2, l_s2, l_v2])
    upper_red2 = np.array([u_h2, u_s2, u_v2])

    # Create masks for both red ranges
    mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
    mask2 = cv2.inRange(hsv, lower_red2, upper_red2)

    # Combine both masks
    mask = mask1 + mask2

    # Apply the mask to the original image
    result = cv2.bitwise_and(image, image, mask=mask)

    # Display results
    cv2.imshow("Original Image", image)
    cv2.imshow("Mask", mask)
    cv2.imshow("Detected Red", result)

    # Press 'ESC' to exit
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def on_trackbar(val):
    global min_area
    min_area = val
    process_image()

def process_image():
    global img, min_area
    
    # Convert image to grayscale and apply threshold
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    
    # Find contours
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    result = img.copy()
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > min_area:
            cv2.drawContours(result, [cnt], -1, (0, 255, 0), 2)
    
    cv2.imshow('Filtered Contours', result)

# Load the image
img = cv2.imread(r"C:\Users\ANUSHRI\Downloads\fruits opencv.jpeg")  # Change 'image.jpg' to your image path
if img is None:
    print("Error: Image not found.")
    exit()

min_area = 100  # Initial area threshold
cv2.namedWindow('Filtered Contours')
cv2.createTrackbar('Min Area', 'Filtered Contours', min_area, 1000, on_trackbar)

process_image()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
